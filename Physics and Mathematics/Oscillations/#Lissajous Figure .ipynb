# Lissajous Figure 

$$
\begin{align}
    x &= A sin(\omega_x t + \delta_1)\\
    y &= B sin(\omega_y t + \delta_2)
\end{align}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
t0 = 0 
t1 = 10 
A = 1
B = 1
pi = np.pi 
n1, n2, n3, n4, n5, n6 =1,1,1,1,1,1 
m1, m2, m3, m4, m5, m6 =2,2,2,2,2,2
d1, d2, d3, d4, d5, d6 =0,0,0,0,0,0  
e1, e2, e3, e4, e5, e6 =0, pi/4, pi/2, 3*pi/4, pi, -pi/2 
```

```python
step = 0.01 # setting the value of step .This will increament t
t = np.arange(t0, t1, step)

#lets create the x and y variables to store the values, x and y will be arrays
size=len(t)
x1=np.empty(size)
y1=np.empty(size)

x2=np.empty(size)
y2=np.empty(size)

x3=np.empty(size)
y3=np.empty(size)

x4=np.empty(size)
y4=np.empty(size)

x5=np.empty(size)
y5=np.empty(size)

x6=np.empty(size)
y6=np.empty(size)
```

```python
#now we will start the actual calculation
i=0
while(i<size):
    x1[i]=A*np.sin(n1*t[i]+d1)
    y1[i]=B*np.sin(m1*t[i]+e1)
    i=i+1

i=0
while(i<size):
    x2[i]=A*np.sin(n2*t[i]+d2)
    y2[i]=B*np.sin(m2*t[i]+e2)
    i=i+1
    
i=0
while(i<size):
    x3[i]=A*np.sin(n3*t[i]+d3)
    y3[i]=B*np.sin(m3*t[i]+e3)
    i=i+1
    
i=0
while(i<size):
    x4[i]=A*np.sin(n4*t[i]+d4)
    y4[i]=B*np.sin(m4*t[i]+e4)
    i=i+1
    
i=0
while(i<size):
    x5[i]=A*np.sin(n5*t[i]+d5)
    y5[i]=B*np.sin(m5*t[i]+e5)
    i=i+1
    
i=0
while(i<size):
    x6[i]=A*np.sin(n6*t[i]+d6)
    y6[i]=B*np.sin(m6*t[i]+e6)
    i=i+1
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(14,10))

plt.subplot(3,2,1)
plt.plot(x1,y1, color="r", linewidth= 1)
plt.title("$\omega_x=1, \quad \omega_y=2, \quad \delta=0$")

plt.subplot(3,2,2)
plt.plot(x2,y2, color="b", linewidth= 1)
plt.title("$\omega_x=1, \quad \omega_y=2, \quad \delta=\pi/4$")

plt.subplot(3,2,3)
plt.plot(x3,y3, color="g", linewidth= 1)
plt.title("$\omega_x=1, \quad \omega_y=2, \quad \delta=\pi/2$")

plt.subplot(3,2,4)
plt.plot(x4,y4, color="black", linewidth= 1)
plt.title("$\omega_x=1, \quad \omega_y=2, \quad \delta=3\pi/4$")

plt.subplot(3,2,5)
plt.plot(x5,y5, color="magenta", linewidth= 1)
plt.title("$\omega_x=1, \quad \omega_y=2, \quad \delta=\pi$")

plt.subplot(3,2,6)
plt.plot(x6,y6, color="orange", linewidth= 1)
plt.title("$\omega_x=1, \quad \omega_y=2, \quad \delta=-\pi/2$")

plt.suptitle("$Lissajous$ $Figure$", fontsize=16)
plt.tight_layout()
plt.show()
```
